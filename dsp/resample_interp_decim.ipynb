# Resampling: Interpolation and Decimation by integer values

A common problem in SDR is resampling of signals. On an SDR device, there's typically a DAC and an ADC running at a fixed speed, e.g. 100 MSps for the USRP N210. In an SDR application, you don't want to deal with these kinds of sampling rates unless absolutely necessary. If your signal is only 250 kHz wide, you don't want more than 250 ksps (complex), or you're wasting resources.

Somewhere between your code and the hardware, you need to change the sampling rate. This is called *resampling*. For this tutorial, let's stick to the case where we interpolate or decimate by integer values, i.e. the output sampling rate $r_{o}$ is an integer multiple of the input sampling rate:

$$r_i = n \cdot r_o \; , \; n \in \mathbb{R}$$

This case is called *interpolation*. *Decimation* is when the sampling rate is reduced by an integer factor:

$$r_i = \frac{r_o}{n} \; , \; n \in \mathbb{R}$$

Only allowing one or the other seems a bit strict, and indeed, there are many other ways to change the sampling rate in a non-integer fashion. But when going from software to hardware, this is quite a common requirement, and in an typical GNU Radio flow graph for example, you want to make sure the final rate produced by your application is indeed within an integer factor of the hardware sampling rate.

But let's get set up and see some signals in action:

```python
%matplotlib inline

# Imports:
import scipy as sp
from scipy import signal
from scipy import fftpack
from scipy import pi
import matplotlib.pyplot as plt 

# Our test signal:
N_x = 2048 # Length of signal in samples
k = sp.arange(N_x) # Time index of x
f_1 = 0.1 # Normalized frequency
f_2 = -.05
f_3 = 0.48
a_1 = 1 # Amplitudes
a_2 = 0.5
a_3 = 0.2
a_N = 0.1
x = a_1 * sp.exp(1j * 2 * pi * f_1 * k) + \
    a_2 * sp.exp(1j * 2 * pi * f_2 * k) + \
    a_3 * sp.exp(1j * 2 * pi * f_3 * k) + \
    a_N * sp.randn(N_x)
N_fft = 1024
P_xx = fftpack.fftshift(signal.welch(x, nfft=N_fft)[1])
f_axis = sp.arange(-.5, .5, 1./N_fft)

# Plot the test signal's spectrum:
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xx)))
plt.title('Test signal')
plt.xlim([-.5,.5])
plt.grid()
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate');
```

This test signal has some nice properties useful for visualizing the following steps. Let's interpret it this way: The center two sinusoids represent a complex (i.e. non-symmetric) signal (our payload). The third sinusoid, at frequency 0.45, is just to cause trouble. All three sinusoids have different amplitudes so we can more easily keep track of them.

Also, we'll have some noise in there just so all frequencies are occupied with at least something.

Let's resample.

## The Theory and Basics

One very obvious problem we need to solve is the fact that when we change the sampling rate, we have more (or less) samples afterwards than we do beforehand. So at some point, we must have a function or a circuitry that discards samples or adds them. A naive approach would be to simply drop $n-1$ out of $n$ samples when decimating by factor $n$, and adding $n-1$ zeros between samples when interpolating (note we can't just add anything else between samples, such as a repetition of the previous sample, because that would be changing the signal. If we just add zeros, we'll definitely and mathematically accurately still have the old signal in there). These operations shall be known as *upsampling* and *downsampling*.

Unless stated otherwise, we'll use a decimation/interpolation factor of 2. What does up- and downsampling do to our signal?

```python
def upsample(x, n):
    return [x[i/n] if ((i%n) == 0) else 0 for i in range(len(x)*n)]

def downsample(x, n):
    return x[::n]
    
x_downsample = downsample(x, 2)
x_upsample = upsample(x, 2)
P_xdxd = fftpack.fftshift(signal.welch(x_downsample, nfft=N_fft)[1])
P_xuxu = fftpack.fftshift(signal.welch(x_upsample, nfft=N_fft)[1])

# Plot the test signal's spectrum:
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xdxd)))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.title('Downsampled Signal')
plt.grid()
plt.xlim([-.5, .5])
plt.figure()
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xuxu)))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.xlim([-.5, .5])
plt.title('Upsampled Signal')
plt.grid()
```

What happened? First of all, we must remember that these plots always show us normalized frequencies -- normalized, that is, to the sampling rate. The span of the x-axis is from the lowest possible frequency representable with the current sampling rate up to the highest. We call this area the 0-th *Nyquist zone*.

However, the sampling rate changed during these operations. Also, we must not forget that the spectrum of a discrete-time signal repeats outside the Nyquist zone: The spectrum from 0.5 to 1.5 (the 1st Nyqust zoen) is the same as from -0.5 to 0.5. 

This explains the upsampled spectrum: By inserting zeros, all we did was double the size of the Nyquist zone and then re-label the x-axis by re-normalizing it to the new sampling rate. So, in addition to our actual signal, we get the *aliases* or *images* from the adjacent Nyquist zones as part of our new signal. We don't want them in there, though. They're messing up our spectrum!

How can we remove these images/aliases? Well, we really only want the signal part around DC, so a low pass filter will help us, right? For the upsampled signal, we simply run the low pass after the upsampling to remove the images (which is why this kind of filter is sometimes called an image rejection filter).

Let's see: Our signal only has components up to a frequency of 0.1 in the original signal (the third component, the annoying third sinusoid, was only there to annoy us). After upsampling, this has dropped to a frequency of 0.05. So, if we design a filter with a stop band frequency of, say 0.1, and a filter transition width of 0.2, we should be fine, right? 

```python
h_interp = signal.firwin(91, .2, width=.2)
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xuxu)))
plt.plot(f_axis, 10 * sp.log10(fftpack.fftshift(signal.freqz(h_interp, worN=N_fft, whole=True)[1])))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.xlim([-.5, .5])
plt.title('Upsampled Signal and filter response overlaid')
plt.grid()
plt.legend(["Upsampled Signal", "Filter Response"])

# Run filter:
x_interp = signal.lfilter(h_interp, [1,], x_upsample)
P_xixi = fftpack.fftshift(signal.welch(x_interp, nfft=N_fft)[1])
plt.figure()
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xixi)))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.xlim([-.5, .5])
plt.ylim([-40, 20])
plt.title('Filtered, Upsampled Signal')
plt.grid()
```

Great! Of course, the images are still there, but severely attenuated. In reality, they will probably be so low compared to the signal and actual receiver noise that they won't be visible any more. We have successfully interpolated a signal!

Decimation is the same, basically, but the filtering must be applied *before* the downsampling, because we want to avoid the aliases in the first place. This means we can do pretty much the same as before, only we need a higher cut-off frequency because in our original signal, frequencies go up to 0.1. However, we want to cut out anything between 0.25 and 0.5, because that's where the aliases come from. So we need a slightly higher cutoff frequency. On y va:

```python
h_decim = signal.firwin(37, .25, width=.1)
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xx)))
plt.plot(f_axis, 10 * sp.log10(fftpack.fftshift(signal.freqz(h_decim, worN=N_fft, whole=True)[1])))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.xlim([-.5, .5])
plt.title('Original Signal and filter response overlaid')
plt.grid()
plt.legend(["Upsampled Signal", "Filter Response"])

# Run filter:
x_decim = signal.lfilter(h_decim, [1,], x)
P_xdxd = fftpack.fftshift(signal.welch(x_decim, nfft=N_fft)[1])

plt.figure()
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xdxd)))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.xlim([-.5, .5])
plt.ylim([-60, 25])
plt.title('Filtered, Original Signal')
plt.grid()

# Now downsample:
x_downsample = downsample(x_decim, 2)
P_xdxd = fftpack.fftshift(signal.welch(x_downsample, nfft=N_fft)[1])

plt.figure()
plt.plot(f_axis, 10 * sp.log10(sp.absolute(P_xdxd)))
plt.xlabel('Normalized frequency')
plt.ylabel('Spectral Power Density Estimate')
plt.xlim([-.5, .5])
plt.ylim([-40, 25])

plt.title('Filtered, Decimated Signal')
plt.grid()
```

We can definitely call this a success! Our original signal is all there, we've pretty aggressively filtered out anything beyond, and the sampling rate was reduced by half. We can tell that the sinusoid originally at frequency 0.1 is now at 0.2.

So is this all there is to resampling? Unfortunately, no. What we did here is theoretically sound, but it lacks practicability. There are two major problems: First, we took advantage of knowledge of the signal before designing our filters. Second, we had no regard for implementation. We simply don't care what our filter does and how it looks like, as long as it does the job. In practice, we often need to design the resampling without any signal knowledge, and have to take processing power limitations into consideration.
